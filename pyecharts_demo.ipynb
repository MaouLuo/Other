```python
import pyecharts
from pyecharts import options as opts
from pyecharts.globals import ThemeType
#print(pyecharts.__version__)
#print(pyecharts.globals.ThemeType)
from faker import Faker
```

```python
bar = pyecharts.charts.Bar()
bar.add_xaxis(['a', 'b', 'c', 'd', 'e'])
bar.add_yaxis('s a', [-12, 23,53,32,12], )
bar.add_yaxis("商家B", [15, 6, 45, 20, 35, 66], category_gap=50)
bar.add_yaxis("商家c", [31, 26, 35, 29, 13, 46])

bar.set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))
# bar.render() # 生成html文件
bar.render_notebook() # 直接在jupyter中渲染
```

```python
f=Faker(locale='zh_CN')
charts = pyecharts.charts.Radar()
charts.add_schema(schema=[opts.RadarIndicatorItem(f.company_prefix(), max_=10),
                  opts.RadarIndicatorItem(f.company_prefix(), max_=10),
                  opts.RadarIndicatorItem(f.company_prefix(), max_=10),
                  opts.RadarIndicatorItem(f.company_prefix(), max_=10),
                  opts.RadarIndicatorItem(f.company_prefix(), max_=10)])
charts.add('销量', data=[[f.random_int(min=3,max=10) for i in range(5)]])
charts.add('产量', data=[[f.random_int(min=3,max=10) for i in range(5)]], linestyle_opts=opts.LineStyleOpts(color="#5CACEE"))

charts.set_global_opts(title_opts=opts.TitleOpts(title="主标题", subtitle="副标题"))
# bar.render() # 生成html文件
charts.render_notebook() # 直接在jupyter中渲染
```
