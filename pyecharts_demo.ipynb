```python
url = 'http://api.k780.com'
data = {
    'weaid': '南山',
    'app': 'weather.future',
    'appkey': tools.r_conf('nowapi', name='appkey'),
    'sign': tools.r_conf('nowapi', name='sign'),
    'format': 'json'
}
resp = requests.post(url, data=data)
jd = json.loads(resp.text)

date_list = [d['days'] for d in jd['result']]
temp_low = [t['temp_low'] for t in jd['result']]
temp_high = [t['temp_high'] for t in jd['result']]

#print(date_list)
```

```python
f=Faker(locale='zh_CN')
charts = pyecharts.charts.Line()

charts.add_xaxis(xaxis_data=date_list)
charts.add_yaxis("High", temp_high, is_smooth=True )
charts.add_yaxis("Low", temp_low, is_smooth=True)

charts.set_global_opts(title_opts=opts.TitleOpts(title="南山天气预报（5-7天）", subtitle="摄氏度"), tooltip_opts=opts.TooltipOpts(trigger="axis"), toolbox_opts=opts.ToolboxOpts(is_show=True) )

# bar.render() # 生成html文件
charts.render_notebook() # 直接在jupyter中渲染
```
